```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
#scatter plot for age and fare columns in the titanic dataset.
```

```python
import plotly.graph_objects as go
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age, y=titanic.fare , mode='markers' ))
fig.show()
```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
tips = sns.load_dataset('tips')
```

```python
import plotly.express as px

tips = px.data.tips()
fig = px.box(tips, x="time", y="tip")

fig.show()
```

```python
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the
#pattern_shape parameter and the "day" column with the color parameter.
```

```python
import plotly.express as px

tips = px.data.tips()
fig = px.histogram(tips, x="sex", y="total_bill", color="day", marginal="rug", pattern_shape="smoker")

fig.show()
```

```python
 # Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
 # Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.
```

```python
import plotly.express as px

iris = px.data.iris()
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")

fig.show()
```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
#A distplot is a histogram with a line drawn over it that shows the estimated probability density function. It is used to visualize the distribution of a dataset.
```

```python
import plotly.express as px

tips = px.data.tips()
fig = sns.distplot(tips["total_bill"])

# Show the plot
fig.show()
```
